```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class UserRoleBean:
    userRoleId = ""
    userRole = ""
    description = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class UserRoleDAO:
    def addUserRoleRecord(self, u=UserRoleBean()):
        try:
     
            sql = "INSERT INTO user_roles(user_role_id, user_role, description, entry_date) VALUES(%s, %s, %s, %s)"
            data = (u.userRoleId, u.userRole, u.description, u.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateUserRoleRecord(self, u=UserRoleBean()):
        try:
            
            sql = "UPDATE user_roles SET user_role=%s, description=%s, entry_date=%s WHERE user_role_id=%s"
            data = (u.userRole, u.description, u.entryDate, u.userRoleId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

            
    def deleteUserRoleRecord(self, userRoleId):
        try:
            
            sql = "DELETE FROM user_roles WHERE user_role_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, (userRoleId,))
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getUserRoleRecord(self, userRoleId):
        try:
            
            sql = "SELECT user_role_id Id, user_role role, description description, entry_date date FROM user_roles WHERE user_role_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (userRoleId,))
            record = cursor.fetchone()
            
            u = UserRoleBean()
            for v in record:
                u.userRoleId = record['Id']
                u.userRole = record['role']
                u.description = record['description']
                u.entryDate = record['date']
            
            return u
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getUserRoleList(self):
        try:
            
            sql = "SELECT user_role_id Id,user_role role, description description, entry_date date FROM user_roles"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            userRoles = []
            for index in range(len(records)):
                userRole = UserRoleBean()
                userRole.userRoleId = records[index]['Id']
                userRole.userRole = records[index]['role']
                userRole.description = records[index]['description']
                userRole.entryDate= records[index]['date']
                
                userRoles.append(userRole)
            
            return userRoles
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
