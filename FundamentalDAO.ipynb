```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class FundamentalBean:
    fundamentalId = ""
    netIncoome = ""
    assetReturn = ""
    cashFlow = ""
    totalAssets = ""
    grossProfit = ""
    turnover = ""
    longTermAssets = ""
    currentAssets = ""
    currentLiabilities = ""
    outstandingShares = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class FundamentalDAO:
    def addFundamentalRecord(self, f=FundamentalBean()):
        try:
     
            sql = "INSERT INTO fundamentals(fundamentals_id, net_income, return_on_assets, cash_flow, total_assets, gross_profit, turnover, long_term_assets, current_assets, current_liabilities, outstanding_shares, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
            data = (f.fundamentalId,f.netIncome, f.assetReturn, f.cashFlow, f.totalAssets, f.grossProfit, f.turnover, f.longTermAssets, f.currentAssets, f.currentLiabilities, f.outstandingShares, f.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateFundamentalRecord(self, f=FundamentalBean()):
        try:
            
            sql = "UPDATE fundamentals SET net_income=%s, return_on_assets=%s, cash_flow=%s, total_assets=%s, gross_profit=%s, turnover=%s, long_term_assets=%s, current_assets=%s, current_liabilities=%s, outstanding_shares=%s, entry_date=%s WHERE fundamentals_id=%s"
            data = (f.netIncome, f.assetReturn, f.cashFlow, f.totalAssets, f.grossProfit, f.turnover, f.longTermAssets, f.currentAssets, f.currentLiabilities, f.outstandingShares, f.entryDate, f.fundamentalId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

            
    def deleteFundamentalRecord(self, fundamentalId):
        try:
            
            sql = "DELETE FROM fundamentals WHERE fundamentals_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, (fundamentalId,))
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getFundamentalRecord(self, fundamentalId):
        try:
            
            sql = "SELECT fundamentals_id Id, net_income income, return_on_assets assets, cash_flow cashflow, total_assets total, gross_profit gross, turnover turnover, long_term_assets longterm, current_assets currentassets, current_liabilities currentliabilities, outstanding_shares shares, entry_date date FROM fundamentals WHERE fundamentals_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (fundamentalId,))
            record = cursor.fetchone()
            
            f = FundamentalBean()
            for v in record:
                f.fundamantalId = record['Id']
                f.netIncome = record['income']
                f.assetReturn = record['assets']
                f.cashFlow = record['cashflow']
                f.totalAssets = record['total']
                f.grossProfit = record['gross']
                f.turnover = record['turnover']
                f.longTermAssets = record['longterm']
                f.currentAssets = record['currentassets']
                f.currentLiabilities = record['currentliabilities']
                f.outstandingShares = record['shares']
                f.entryDate = record['date'] 
                
            return f
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def getFundamentalList(self):
        try:
            
            sql = "SELECT fundamentals_id Id, net_income income, return_on_assets assets, cash_flow cashflow, total_assets total, gross_profit gross, turnover turnover, long_term_assets longterm, current_assets currentassets, current_liabilities currentliabilities, outstanding_shares shares, entry_date date FROM fundamentals"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            fundamentals = []
            for index in range(len(records)):
                fundamental = FundamentalBean()
                fundamental.fundamantalId = records[index]['Id']
                fundamental.netIncome = records[index]['income']
                fundamental.assetReturn = records[index]['assets']
                fundamental.cashFlow = records[index]['cashflow']
                fundamental.totalAssets = records[index]['total']
                fundamental.grossProfit = records[index]['gross']
                fundamental.turnover = records[index]['turnover']
                fundamental.longTermAssets = records[index]['longterm']
                fundamental.currentAssets = records[index]['currentassets']
                fundamental.currentLiabilities = records[index]['currentliabilities']
                fundamental.outstandingShares = records[index]['shares']
                fundamental.entryDate = records[index]['date'] 
            
                
                fundamentals.append(fundamental)
            
            return fundamentals
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
