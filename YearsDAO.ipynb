```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class YearBean:
    yearId = ""
    year = ""
```

```python
class YearDAO:
    def addYearRecord(self, y=YearBean()):
        try:
     
            sql = "INSERT INTO fin_years(year_id, year) VALUES(%s, %s)"
            data = (y.yearId, y.year,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateYearRecord(self, y=YearBean()):
        try:
            
            sql = "UPDATE fin_years SET year=%s WHERE year_id=%s"
            data = (y.year, y.yearId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()

            
    def deleteYearRecord(self, yearId):
        try:
            
            sql = "DELETE FROM fin_years WHERE year_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, (yearId,))
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getYearRecord(self, yearId):
        try:
            
            sql = "SELECT year_id Id, year year FROM fin_years WHERE year_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (yearId,))
            record = cursor.fetchone()
            
            y = YearBean()
            for v in record:
                y.yearId = record['Id']
                y.year = record['year']
            
            return y
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getYearList(self):
        try:
            
            sql = "SELECT year_id Id, year year FROM fin_years"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            years = []
            for index in range(len(records)):
                year = YearBean()
                year.yearId = records[index]['Id']
                year.year = records[index]['year']

                
                years.append(year)
            
            return years
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
