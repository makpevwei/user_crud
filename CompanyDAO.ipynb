```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class CompanyBean:
    companyId = ""
    companyName = ""
    symbol = ""
    business = ""
    address = ""
    email = ""
    mobileNumber = ""
    dateListed = datetime.now().strftime('%Y-%m-%d')
    dateIncorporated = datetime.now().strftime('%Y-%m-%d')
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class CompanyDAO:
    def addCompanyRecord(self, c=CompanyBean()):
        try:
     
            sql = "INSERT INTO companies(company_id, company_name, symbol, nature_of_business, address, email, mobile_no, date_listed, date_of_incorporation, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
            data = (c.companyId,c.companyName, c.symbol, c.business, c.address, c.email, c.mobileNumber, c.dateListed, c.dateIncorporated, c.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateCompanyRecord(self, c=CompanyBean()):
        try:
            
            sql = "UPDATE companies SET company_name=%s, symbol=%s, nature_of_business=%s, address=%s, email=%s, mobile_no=%s, date_listed=%s, date_of_incorporation=%s, entry_date=%s WHERE company_id=%s"
            data = (c.companyName, c.symbol, c.business, c.address, c.email, c.mobileNumber, c.dateListed, c.dateIncorporated, c.entryDate, c.companyId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

            
    def deleteCompanyRecord(self, companyId):
        try:
            
            sql = "DELETE FROM companies WHERE company_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, (companyId,))
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getCompanyRecord(self, companyId):
        try:
            
            sql = "SELECT company_id Id, company_name name, symbol symbol, nature_of_business business, address address, email email, mobile_no mobile, date_listed listed, date_of_incorporation incorporation, entry_date date FROM companies WHERE company_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (companyId,))
            record = cursor.fetchone()
            
            c = CompanyBean()
            for v in record:
                c.companyId = record['Id']
                c.companyName = record['name']
                c.symbol = record['symbol']
                c.business = record['business']
                c.address = record['address']
                c.email = record['email']
                c.mobileNumber = record['mobile']
                c.dateListed = record['listed']
                c.dateIncorporated = record['incorporation']
                c.entryDate = record['date']
            
            return c
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def getCompanyList(self):
        try:
            
            sql = "SELECT company_id Id, company_name name, symbol symbol, nature_of_business business, address address, email email, mobile_no mobile, date_listed listed, date_of_incorporation incorporation, entry_date date FROM companies"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            companies = []
            for index in range(len(records)):
                company = CompanyBean()
                company.companyId = records[index]['Id']
                company.companyName = records[index]['name']
                company.symbol = records[index]['symbol']
                company.business = records[index]['business']
                company.address = records[index]['address']
                company.email = records[index]['email']
                company.mobileNumber = records[index]['mobile']
                company.dateListed = records[index]['listed']
                company.dateIncorporated = records[index]['incorporation']
                company.entryDate = records[index]['date']
            
                
                companies.append(company)
            
            return companies
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
