```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class QuarterBean:
    quarterId = ""
    quarter = ""
```

```python
class QuarterDAO:
    def addQuarterRecord(self, q=QuarterBean()):
        try:
     
            sql = "INSERT INTO fin_quarters(quarter_id, quarters) VALUES(%s, %s)"
            data = (q.quarterId, q.quarter,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateQuarterRecord(self, q=QuarterBean()):
        try:
            
            sql = "UPDATE fin_quarters SET quarters=%s WHERE quarter_id=%s"
            data = (q.quarter, q.quarterId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()

            
    def deleteQuarterRecord(self, quarterId):
        try:
            
            sql = "DELETE FROM fin_quarters WHERE quarter_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, (quarterId,))
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getQuarterRecord(self, quarterId):
        try:
            
            sql = "SELECT quarter_id Id, quarters quarter FROM fin_quarters WHERE quarter_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (quarterId,))
            record = cursor.fetchone()
            
            q = QuarterBean()
            for v in record:
                q.quarterId = record['Id']
                q.quarter = record['quarter']
            
            return q
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getQuarterList(self):
        try:
            
            sql = "SELECT quarter_id Id, quarters quarter FROM fin_quarters"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            quarters = []
            for index in range(len(records)):
                quarter = QuarterBean()
                quarter.quarterId = records[index]['Id']
                quarter.quarter = records[index]['quarter']

                
                quarters.append(quarter)
            
            return quarters
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
