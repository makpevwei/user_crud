```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class UserBean:
    userId = ""
    userRoleId = ""
    firstName = ""
    lastame = ""
    gender = ""
    mobileNumber = ""
    email = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class UserDAO:
    def addUserRecord(self, u=UserBean()):
        try:
     
            sql = "INSERT INTO users(user_id, user_role_id, first_name, last_name, gender, mobile_no, email, entry_date) VALUES(%s, %s, %s, %s, %s, %s, %s, %s)"
            data = (u.userId, u.userRoleId, u.firstName, u.lastName, u.gender, u.mobileNumber, u.email, u.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateUserRecord(self, u=UserBean()):
        try:
            
            sql = "UPDATE users SET user_role_id=%s, first_name=%s, last_name=%s, gender=%s,  mobile_no=%s,  email=%s, entry_date=%s WHERE user_id=%s"
            data = (u.userRoleId, u.firstName, u.lastName, u.gender, u.mobileNumber, u.email, u.entryDate, u.userId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

            
    def deleteUserRecord(self, userId):
        try:
            
            sql = "DELETE FROM users WHERE user_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, (userId,))
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getUserRecord(self, userId):
        try:
            
            sql = "SELECT user_id Id, user_role_id roleId, first_name first, last_name last, gender gender, mobile_no mobile, email email, entry_date date FROM users WHERE user_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (userId,))
            record = cursor.fetchone()
            
            u = UserBean()
            for v in record:
                u.userId = record['Id']
                u.userRoleId = record['roleId']
                u.firstName = record['first']
                u.lastName = record['last']
                u.gender = record['gender']
                u.mobileNumber = record['mobile']
                u.email = record['email']
                u.entryDate = record['date']
            
            return u
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getUserList(self):
        try:
            
            sql = "SELECT user_id Id, user_role_id roleId, first_name first, last_name last, gender gender, mobile_no mobile, email email, entry_date date FROM users"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            users = []
            for index in range(len(records)):
                user = UserBean()
                user.userId = records[index]['Id']
                user.userRoleId = records[index]['roleId']
                user.firstName = records[index]['first']
                user.lastName = records[index]['last']
                user.gender = records[index]['gender']
                user.mobileNumber= records[index]['mobile']
                user.email = records[index]['email']
                user.entryDate = records[index]['date']
                
                users.append(user)
            
            return users
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
