```python
import pymysql
from db_config import mysql
from datetime import datetime
```

```python
class ApplicationBean:
    appId = ""
    userId = ""
    appName = ""
    entryDate = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
```

```python
class ApplicationDAO:
    def addApplicationRecord(self, a=ApplicationBean()):
        try:
     
            sql = "INSERT INTO applications(app_id, user_id, app_name, entry_date) VALUES(%s, %s, %s, %s)"
            data = (a.appId, a.userId, a.appName, a.entryDate,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
            
    def updateApplicationRecord(self, a=ApplicationBean()):
        try:
            
            sql = "UPDATE applications SET user_id=%s, app_name=%s, entry_date=%s WHERE app_id=%s"
            data = (a.userId, a.appName, a.entryDate, a.appId,)
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, data)
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()

            
    def deleteApplicationRecord(self, appId):
        try:
            
            sql = "DELETE FROM applications WHERE app_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor()
            cursor.execute(sql, (appId,))
            conn.commit()
            return 0
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            

    def getApplicationRecord(self, appId):
        try:
            
            sql = "SELECT app_id Id, user_id userId, app_name name, entry_date date FROM applications WHERE app_id=%s"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql, (appId,))
            record = cursor.fetchone()
            
            a = ApplicationBean()
            for v in record:
                a.appId = record['Id']
                a.userId = record['userId']
                a.appName = record['name']
                a.entryDate = record['date']
            
            return a
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
            
    def getApplicationList(self):
        try:
            
            sql = "SELECT app_id Id, user_id userId, app_name name, entry_date date FROM applications"
            conn = mysql.connect()
            cursor = conn.cursor(pymysql.cursors.DictCursor)
            cursor.execute(sql)
            records = cursor.fetchall()
            
            applications = []
            for index in range(len(records)):
                application = ApplicationBean()
                application.appId = records[index]['Id']
                application.userId = records[index]['userId']
                application.appName = records[index]['name']
                application.entryDate = records[index]['date']
                
                applications.append(application)
            
            return applications
        except Exception as e:
            print(e)
        finally:
            cursor.close() 
            conn.close()
```
